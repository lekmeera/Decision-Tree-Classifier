```python
#importing files and reading dataset

import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.columns
```

```python
#using only four attributes to predict 1.Pclass 2.Sex 3.Age 4.Fare so dropping the rest attributes

df1 = df.drop(columns=['PassengerId', 'Name','SibSp',
       'Parch', 'Ticket', 'Cabin', 'Embarked'],axis='columns')
df1
```

```python
# to find the null values

df1.isnull().sum()
```

```python
#to convert categorical data into numerical

from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
df1['Sex_n'] = le_sex.fit_transform(df1['Sex'])
```

```python
df1
```

```python
#splitting attributes as input and target

target = df1['Survived']
```

```python
inputs = df1.drop(columns=['Sex','Survived'],axis='columns')
inputs
```

```python
# finding imputation using mean() method

inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
target.dtypes
```

```python
#training and testing

from sklearn.model_selection import train_test_split
x = inputs
y = target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
#finding accuracy using Decision tree classifier

from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
